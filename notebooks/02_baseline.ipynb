Goal:
- Build the simplest model (Logistic Regression)
- Use minimal preprocessing
- Get baseline PR-AUC & ROC-AUC
- This will be compared with optimized models later

```python
from sklearn.pipeline import Pipeline
from sklearn.metrics import roc_auc_score, average_precision_score

from src.data_prep import load_raw_data, train_val_test_split, build_preprocessor
from src.models import get_baseline_model
from src.utils import evaluate_probabilities, print_classification_info
```

```python
df = load_raw_data()
X_train, X_val, X_test, y_train, y_val, y_test = train_val_test_split(df)
```

```python
preprocessor = build_preprocessor()
model = get_baseline_model()

baseline_pipe = Pipeline([
    ("preprocessor", preprocessor),
    ("model", model)
])

baseline_pipe.fit(X_train, y_train)
```

```python
y_val_proba = baseline_pipe.predict_proba(X_val)[:, 1]
metrics = evaluate_probabilities(y_val, y_val_proba)
metrics
```

```python
y_pred = (y_val_proba >= 0.5).astype(int)
print_classification_info(y_val, y_pred)
```
